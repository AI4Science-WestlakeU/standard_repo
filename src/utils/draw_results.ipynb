# 1. Draw curve for different \lambda

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def get_mean_from_str(str):
    # print(str)
    str_split=str.split()[0]
    mean=float(str_split)
    return mean
```

```python
results=pd.read_excel("/zhangtao/project/inverse_design/results/results_1d_inverse_design_lambda.xlsx",header=None)
# results_array=results.astype(str).values
# results
results_array=np.zeros((11,9))
for i in range(9):
    for j in range(11):
        print(results[i][j])
        if i==0:
            results_array[j][i]=float(results[i][j])
        else:
            results_array[j][i]=get_mean_from_str(results[i][j])
```

```python
for i in range(11):
    for j in range(9):
        print(results_array[i][j])
```

```python
results_array_log=results_array
results_array_log[:,0]=np.log10(results_array[:,0])
results_array_log
```

```python
print(results_array_log[:,0])
```

```python
fontsize=18
plt.figure()
plt.plot()

plt.plot( results_array_log[:,0],results_array_log[:,1], label='design obj of 24 steps', marker='o',color='blue')
plt.plot( results_array_log[:,0],results_array_log[:,3], label='design obj of 34 steps', marker='*',color='green')
plt.plot( results_array_log[:,0],results_array_log[:,5], label='design obj of 44 steps', marker='^',color="red")
plt.plot( results_array_log[:,0],results_array_log[:,7], label='design obj of 54 steps', marker='s',color="purple")


plt.legend()
# plt.title(r'design obj of different $\lambda$')
plt.xlabel(r'$\log_{10}\lambda$',fontsize=fontsize)
plt.ylabel(r'design obj',fontsize=fontsize)
plt.tick_params(labelsize=12)
plt.ylim(0,0.35)
plt.savefig('/zhangtao/project/inverse_design/results/inverse_design_diffusion/design_obj_of_different_lambda.pdf')
```

```python
fontsize=18
plt.figure()
plt.plot()

plt.plot( results_array_log[:,0],results_array_log[:,2], label='MAE of 24 steps', marker='o',color='blue')
plt.plot( results_array_log[:,0],results_array_log[:,4], label='MAE of 34 steps', marker='*',color='green')
plt.plot( results_array_log[:,0],results_array_log[:,6], label='MAE of 44 steps', marker='^',color="red")
plt.plot( results_array_log[:,0],results_array_log[:,8], label='MAE of 54 steps', marker='s',color="purple")


plt.legend()
# plt.title(r'MAE of different $\lambda$')
plt.xlabel(r'$\log_{10}\lambda$',fontsize=fontsize)
plt.ylabel(r'MAE',fontsize=fontsize)
plt.tick_params(labelsize=12)
plt.ylim(0,0.25)
plt.savefig('/zhangtao/project/inverse_design/results/inverse_design_diffusion/MAE_of_different_lambda.pdf')
```
